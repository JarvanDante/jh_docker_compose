### 构造要读取的源数据库，源表

```python
#源数据库
source_database="db_hc_uat_report"
#源表
source_tables=["report_hc_group","report_department_employee","report_d_product","report_d_regiment","report_hc_budget_summary","report_d_product"]

#生成查询语句
for index,value in enumerate(source_tables):
    query="show create table "+source_database+"."+value
    print(query)
    df1=pd.read_sql(query,conn)
    if index==0:
        df=df1
    else:
        df=pd.concat([df,df1],ignore_index=True)
df
```

### 查看原始建表语句

```python
df["Create Table"][0]
```

### 修改生成新的建表语句

```python
#导入正则模块
import re
for i in range(df.shape[0]):
    print(i)
    #找出多余的ddl
    replace_str_list=re.findall("\)\ (.*?) COMMENT", df["Create Table"][i])+ re.findall("text (.*?) COMMENT", df["Create Table"][0])
    
    # 将列表转换成集合并排序
    replace_str_list = set(replace_str_list)
    replace_str_list = sorted(replace_str_list,key=lambda s: len(s),reverse=True)
    
    #替换字段类型中的（）和数值：varchar(100) -- varchar
    df["Create Table"][i]=re.sub('\((\d+)\)', '', df["Create Table"][i])
    
    #替换字段类型
    df["Create Table"][i]=df["Create Table"][i].replace('varchar', 'string').replace('text', 'string')
    df["Create Table"][i]=df["Create Table"][i].replace('smallint','int').replace('float', 'double')
    
    for j in replace_str_list:
        #去除多余的ddl
        df["Create Table"][i]=df["Create Table"][i].replace(j,"")

#查看一下是否修改成功
df["Create Table"][0]
```
